```python
# problem 1.7.1
def myfilter(L,num):
    return [x for x in L if x % num == 0]
print(myfilter([x for x in range(10)], 3))
```

```python
# problem 1.7.2
def mylist(L):
    return[[y for y in range(1,x+1,1)] for x in L]
print(mylist([1,3,9]))
```

```python
# problem 1.7.3 (funtional composition)

def composition(A,B):
    return({key1:value2 for (key1, value1) in A.items() for (key2, value2) in B.items() if value1 == key2})
    
A = {'01':3, '02':5, '03':10}
B = {3:'100 Euro', 5:'Green blanket', 10:'Nothing'}

print(composition(A,B))
print(A.get('01'))
```

```python
# problem 1.7.4, 1.7.5, 1.7.6, 1.7.7, 1.7.8
# each function refers to an operation. the initializing of the result
# container can me done smartly by using the identity-element i of this
# operation such that i op x = x
L =  [12,2,3,4,5]
S = ['er',' ','geht',' ','weg!']
X = [{1,3,5},{0,9},{8,1,10}]
def my_sum(L):
    res = 0
    for i in range(len(L)):res += L[i] 
    return res

def my_product(L):
    res = L[0]
    for i in range(len(L)-1):res = res * L[i+1] 
    return res

def my_min(L):
    res = L[0]
    for i in range(len(L)-1):
        if L[i+1] < res:
            res = L[i+1]
    return res

def my_concat(S):
    res = ''
    for i in range(len(L)): res = res + S[i]
    return res

def my_union(X):
    res = set()
    for i in range(len(X)): res = res | X[i]
    return res

print(my_sum(L))
print(my_product(L))
print(my_min(L))
print(my_concat(S))
print(my_union(X))
```

```python
# 1.7.12

def linear_transform(a,b,z):
    return (a*z+b)

def get_transformations(a,b,L):
    liste = [linear_transform(a,b,number) for number in L]
    plt.arrow(0,0,a.real,a.imag, head_width = 0.3,)
    plt.arrow(0,0,b.real,b.imag, head_width = 0.3)
    plt.scatter(L[0].real, L[0].imag, s = 100, color = 'red')
    plt.scatter(liste[0].real, liste[0].imag, s = 100, color = 'blue')
    plt.xlim([-15,15])
    plt.ylim([-15,15])
    plt.show()
    return liste
    
a = -2j
b = 3 + 1j
L = [2+2j,0]
get_transformations(a,b,L)
```
